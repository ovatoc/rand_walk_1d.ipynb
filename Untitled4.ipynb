```python
import numpy as np
import matplotlib.pyplot as plt

Nsteps = 100
Nreps_list = [50, 100, 200, 400, 800]

plt.figure(figsize=(7,5))
for Nrep in Nreps_list:
    # arah acak: 0=N, 1=E, 2=S, 3=W
    directions = np.random.randint(0, 4, size=(Nrep, Nsteps))
    dx = np.where(directions == 1, 1, np.where(directions == 3, -1, 0))
    dy = np.where(directions == 0, 1, np.where(directions == 2, -1, 0))

    x = np.cumsum(dx, axis=1)
    y = np.cumsum(dy, axis=1)
    r = np.sqrt(x**2 + y**2)

    mean_distance = np.mean(r, axis=0)
    plt.plot(range(1, Nsteps+1), mean_distance, label=f'Nrep={Nrep}')

plt.xlabel('Steps')
plt.ylabel('Mean Distance')
plt.title('Mean Distance (2D Random Walk)')
plt.legend()
plt.grid(True)
plt.show()
```

Penjelasan:


*   Semakin besar Nrep, hasil rata-rata semakin stabil dan halus.
*   Nilai mean distance meningkat seiring bertambahnya langkah, dan kira-kira proporsional dengan Nsteps (akar jumlah langkah).
*   Ini sesuai teori random walk 2D bahwa jarak rata-rata tumbuh dengan akar waktu.

```python
import numpy as np
import matplotlib.pyplot as plt

Nsteps = 100
Nreps_list = [50, 100, 200, 400, 800]

plt.figure(figsize=(7,5))
for Nrep in Nreps_list:
    directions = np.random.randint(0, 4, size=(Nrep, Nsteps))
    dx = np.where(directions == 1, 1, np.where(directions == 3, -1, 0))
    dy = np.where(directions == 0, 1, np.where(directions == 2, -1, 0))

    x = np.cumsum(dx, axis=1)
    y = np.cumsum(dy, axis=1)
    r2 = x**2 + y**2

    mean_squared_distance = np.mean(r2, axis=0)
    plt.plot(range(1, Nsteps+1), mean_squared_distance, label=f'Nrep={Nrep}')

plt.xlabel('Steps')
plt.ylabel('Mean Squared Distance')
plt.title('Mean Squared Distance (2D Random Walk)')
plt.legend()
plt.grid(True)
plt.show()
```

Penjelasan

- Mean squared distance (r) meningkat secara linier terhadap jumlah langkah (Nsteps).

- Ini menandakan difusi isotropik di bidang 2-D.

- Semakin besar Nrep, hasil semakin mendekati garis teoritis
