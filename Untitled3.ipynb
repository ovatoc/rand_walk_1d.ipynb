```python
import random as rnd
import matplotlib.pyplot as plt

Nsteps = 20
x_mean = []

for Nrep in [1e2, 1e3, 1e4, 1e5, 1e6]:
    Nrep = int(Nrep)
    x_av = []
    for step in range(Nsteps):
        x = 0
        total = 0
        for i in range(Nrep):
            x += rnd.choice([-1, 1])
            total += x
        x_av.append(total / Nrep)
    x_mean.append(x_av)

for i, Nrep in enumerate([1e2, 1e3, 1e4, 1e5, 1e6]):
    plt.plot(range(Nsteps), x_mean[i], label=f'Nrep={int(Nrep)}')

plt.xlabel('steps')
plt.ylabel('mean distance')
plt.legend()
plt.show()
```

Penjelasan hasil:
*   Saat Nrep (jumlah pengulangan) meningkat, kurva menjadi lebih halus dan mendekati 0.
*   Ini karena semakin banyak percobaan, efek acak saling menetralkan (hukum bilangan besar).
*   Jadi, mean distance → 0 ketika Nrep besar.

Referensi singkat:

Konsep ini mengikuti teori random walk symmetric di mana pergerakan kanan dan kiri memiliki peluang yang sama, sehingga posisi rata-rata setelah banyak percobaan akan mendekati nol (Feller, An Introduction to Probability Theory, 1968).

```python
import random as rnd
import matplotlib.pyplot as plt

Nsteps = 120
x2_mean = []

for Nrep in [25, 50, 100, 200, 400]:
    Nrep = int(Nrep)
    x2_av = []
    for step in range(Nsteps):
        total = 0
        for i in range(Nrep):
            x = 0
            for j in range(step):
                x += rnd.choice([-1, 1])
            total += x**2
        x2_av.append(total / Nrep)
    x2_mean.append(x2_av)

for i, Nrep in enumerate([25, 50, 100, 200, 400]):
    plt.plot(range(Nsteps), x2_mean[i], label=f'Nrep={Nrep}')

plt.xlabel('steps')
plt.ylabel('mean squared distance')
plt.legend()
plt.show()
```

Penjelasan hasil:

Mean squared distance meningkat proporsional terhadap jumlah langkah (Nstep).
Artinya, semakin jauh melangkah, semakin besar penyebaran posisi partikel.

Semakin besar Nrep, grafik semakin halus dan hasilnya semakin mendekati nilai teoritis ⟨𝑥2⟩=𝑁𝑠𝑡𝑒𝑝

🔹 Referensi singkat:

Fenomena ini merupakan dasar dari difusi Brownian, di mana ⟨x2⟩=2Dt (Einstein, 1905).
